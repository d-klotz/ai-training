```python
import cv2
import matplotlib.pyplot as plt

# Load image
image = cv2.imread('data/me.jpg')

# convert image from BGR to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display image
plt.imshow(image_rgb)
plt.axis('off') #disables the axis
plt.show()
```

```python
# Convert to grayscale
img_path = 'data/me.jpg'
image = cv2.imread(img_path)
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# display grayscale image
plt.imshow(image_gray, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Resize the image
width = 800
height = 600

dimensions = (width, height)
imagem_redimensionada = cv2.resize(image_gray, dimensions, interpolation=cv2.INTER_AREA)

plt.imshow(imagem_redimensionada, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Smoothing images -> We can use this technique when we have text on the image that is not very clear

smoothed_image = cv2.GaussianBlur(image_gray, (15, 15), 0)

smoothed_image_rgb = cv2.cvtColor(smoothed_image, cv2.COLOR_GRAY2RGB)

plt.imshow(smoothed_image_rgb)
plt.axis('off')
plt.show()
```

```python
# Detect edges in the image

edges = cv2.Canny(image_gray, 100, 200)

plt.imshow(edges, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Create forms on an image
initial_rectangle = (450, 200)
end_rectangle = (950, 900)
color = (255, 0, 0)
thickness = 2

image_with_rectangle = cv2.rectangle(image_gray.copy(), initial_rectangle, end_rectangle, color, thickness)

image_with_rectangle_rgb = cv2.cvtColor(image_with_rectangle, cv2.COLOR_GRAY2RGB)

plt.imshow(image_with_rectangle_rgb)
plt.axis('off')
plt.show()
```

```python
# saving the image
path_to_save = 'data/image_processed.jpg'
cv2.imwrite(path_to_save, image_with_rectangle)
```
